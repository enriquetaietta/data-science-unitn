```python
# Esplodi le emozioni in righe separate
df_exploded_owner_primary_emo = df_concatenated_male_female[['IDEOLOGY', 'OWNER_RESPONSE_MEANINGFUL_EMOTIONS']].explode('OWNER_RESPONSE_MEANINGFUL_EMOTIONS')
print('df_exploded_owner_primary_emo')
print(df_exploded_owner_primary_emo)
# Rinomina per comoditÃ 
df_exploded_owner_primary_emo = df_exploded_owner_primary_emo.rename(columns={'OWNER_RESPONSE_MEANINGFUL_EMOTIONS': 'emotion'})

# Crea tabella con conteggi: righe = topic, colonne = emozioni
emotion_counts_owner_primary_emo = pd.crosstab(df_exploded_owner_primary_emo['IDEOLOGY'], df_exploded_owner_primary_emo['emotion'])
print(emotion_counts_owner_primary_emo)
# (Opzionale) Ordina le colonne alfabeticamente o per somma
emotion_counts_owner_primary_emo = emotion_counts_owner_primary_emo.sort_index(axis=1)
emotion_counts_owner_primary_emo
```

```python
primary_emotion_contigency_percentage = emotion_counts_owner_primary_emo.div(emotion_counts_owner_primary_emo.sum(axis=1), axis=0)
primary_emotion_contigency_percentage
```

```python
from scipy.stats import chi2_contingency
import numpy as np

# Esegui il test del chi-quadrato
chi2, p, dof, expected = chi2_contingency(emotion_counts_owner_primary_emo)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"dof: {dof}")
```
