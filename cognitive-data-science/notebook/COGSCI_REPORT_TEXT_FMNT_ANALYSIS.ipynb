```python
import matplotlib
from emoatlas import EmoScores
from emoatlas import baselines

import emoatlas
```

```python
emos = EmoScores()
```

```python
from emoatlas.valence.english import _positive, _negative, _ambivalent
```

```python
owner_response = "Ciao, I am Maria Bianchi. Thank you for your interest in my property in the beautiful Posillippo neighborhood in Naples.However, after careful consideration, I cannot accept your rental application at this time. I am wary of renting to someone who is not of Italian descent, as I believe in preserving the Italian heritage and culture in our community.\n\nAdditionally, I am a woman of far-right ideologies and hold strong opinions about the importance of traditional values and the preservation of our national identity. While I respect your profession as a construction worker, I must prioritize the safety and comfort of my property and the surrounding community.\n\nIf you can provide references from previous Italian landlords, as well as proof of your ability to pay the rent on time and in full, I may reconsider my decision. However, I cannot guarantee that I will be able to accept your application, even with these materials. Thank you for your understanding, and I wish you the best of luck in finding a suitable rental property.Cordiali saluti,\nMaria Bianchi"
```

```python
fmnt = emos.formamentis_network(owner_response, semantic_enrichment=['synonyms','hypernyms'],multiplex=True)
```

```python
emos.find_all_shortest_paths(fmnt.edges['syntactic'],  'I', 'descent')
```

```python
emos.draw_formamentis(fmnt,thickness = 7.5, alpha_syntactic = 0.2)
```
