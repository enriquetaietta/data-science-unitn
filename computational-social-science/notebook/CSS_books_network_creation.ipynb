```python
import pandas as pd
```

```python
# CSV relativo ai dati estratti dalla pagina bestsellers nella categoria 'politica'

book_csv = pd.read_csv('book_category_508811031.csv', header=None, delimiter='|')
```

```python
# Recupero dati dal .csv relativo alle review presenti per ciascun libro, con anche il valore relativo alle persone che hanno trovato utile la review

books_reviews_csv = pd.read_csv("product_reviews_comments.csv", header=None, delimiter='|')
```

```python
# Creazione grafo/network tramite networkx
## Caratteristiche
### Nodo type = X -> Book (book_code)
### Nodo type = Y -> Review (book_code+'RW'+review_index)
### Nodo type = Z -> Useful (Utenti che hanno trovato utile la review)  (book_code+'RW'+review_index+'UF'+n)
### Edge type = XY -> Relazion 1 a n tra book e reviews
### Edge type = YZ -> Relazione 1 a n tra review e numero di persone che hanno trovato utile la recensione

import networkx as nx
import matplotlib.pyplot as plt

G = nx.MultiGraph()

for index, row in book_csv.iterrows():
  G.add_node(row[3], type='X')
  print(row[3])
  for index_review, review in books_reviews_csv[books_reviews_csv[0]== row[3]].iterrows():
    G.add_node(row[3]+"RW"+str(index_review), type='Y')
    G.add_edge(row[3], row[3]+"RW"+str(index_review), relation='XY')
    print(row[3]+"RW"+str(index_review))
    try:
      for n in range(0, int(review[4])):
        print(row[3]+"RW"+str(index_review)+"UF"+str(n))
        G.add_node(row[3]+"RW"+str(index_review)+"UF"+str(n), type='Z')
        G.add_edge(row[3]+"RW"+str(index_review), row[3]+"RW"+str(index_review)+"UF"+str(n), relation='YZ')
    except Exception as e:
      print(e)
      print(row)
      print(index)

```

```python
# Export nodes to CSV
nodes_data = [{'Id': node, 'Type': G.nodes[node]['type']} for node in G.nodes()]
nodes_df = pd.DataFrame(nodes_data)
nodes_df.to_csv('nodes_v3.csv', index=False)

# Export edges to CSV
edges_data = [{'Source': u, 'Target': v, 'Relation': G[u][v][0]['relation']} for u, v in G.edges()]
edges_df = pd.DataFrame(edges_data)
edges_df.to_csv('edges_v3.csv', index=False)
```
