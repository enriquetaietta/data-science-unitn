```python
import networkx as nx
import matplotlib.pyplot as plt

# Creazione di un MultiGraph
G = nx.MultiGraph()
G.add_node('TOPIC0', type='T')
G.add_node('TOPIC1', type='T')
G.add_node('TOPIC2', type='T')
G.add_node('TOPIC3', type='T')
G.add_node('TOPIC4', type='T')
G.add_node('TOPIC5', type='T')
G.add_node('TOPIC6', type='T')
G.add_node('TOPIC7', type='T')
G.add_node('TOPIC8', type='T')
G.add_node('TOPIC9', type='T')
for index, book_code in book_csv.iterrows():
  G.add_node(book_code[3], type='X')
  print(book_code[3])
  for review_index,  review in books_reviews_json[books_reviews_json["book_code"]== book_code[3]].iterrows():
    print(review)
    print(review_index)
    print(review["index"])
    G.add_node(book_code[3]+"RW"+str(review["index"]), type='Y')
    G.add_edge(book_code[3], book_code[3]+"RW"+str(review["index"]), relation='XY')
    print(book_code[3]+"RW"+str(review["index"]))
    try:
      for n in range(0, int(review[5])):
        print(book_code[3]+"RW"+str(review["index"])+"UF"+str(n))
        G.add_node(book_code[3]+"RW"+str(review["index"])+"UF"+str(n), type='Z')
        G.add_edge(book_code[3]+"RW"+str(review["index"]), book_code[3]+"RW"+str(review["index"])+"UF"+str(n), relation='YZ')
    except Exception as e:
      print(e)
      print(book_code)
      print(index)

    # Inserisco i nodi-archi relativi ai topic
    if (review["topic_num"] == 0):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC0', relation='YT')
    elif (review["topic_num"] == 1):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC1', relation='YT')
    elif (review["topic_num"] == 2):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC2', relation='YT')
    elif (review["topic_num"] == 3):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC3', relation='YT')
    elif (review["topic_num"] == 4):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC4', relation='YT')
    elif (review["topic_num"] == 5):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC5', relation='YT')
    elif (review["topic_num"] == 6):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC6', relation='YT')
    elif (review["topic_num"] == 7):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC7', relation='YT')
    elif (review["topic_num"] == 8):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC8', relation='YT')
    elif (review["topic_num"] == 9):
      G.add_edge(book_code[3]+"RW"+str(review["index"]), 'TOPIC9', relation='YT')
```

```python
from collections import defaultdict

# Assuming `G` is a graph with node attribute `type`
path_count = defaultdict(lambda: defaultdict(int))

for y in [n for n, d in G.nodes(data=True) if d.get('type') == 'Y']:
    x_neighbors = [n for n in G.neighbors(y) if G.nodes[n].get('type') == 'X']
    t_neighbors = [n for n in G.neighbors(y) if G.nodes[n].get('type') == 'T']

    for x in x_neighbors:
        for t in t_neighbors:
            path_count[x][t] += 1  # X–Y–T path via this review
```

```python
for key, value in path_count.items():
  print(sorted(path_count[key].items(), key=lambda x: -x[1]))

```

```python
sum_path_count = 0
for key, value in path_count.items():
  print(key)
  print(path_count[key]['TOPIC0'])
  sum_path_count += path_count[key]['TOPIC0']

print(sum_path_count)
```

```python
normalized_connectivity = {}

for x, t_counts in path_count.items():
    total = sum(t_counts.values())
    if total == 0:
        continue
    normalized_connectivity[x] = {t: count / total for t, count in t_counts.items()}
```

```python
y_importance = {}
for y in [n for n, d in G.nodes(data=True) if d.get('type') == 'Y']:
    z_neighbors = [n for n in G.neighbors(y) if G.nodes[n].get('type') == 'Z']
    y_importance[y] = len(z_neighbors)  # or another scoring function
```

```python
from collections import defaultdict

weighted_path_count = defaultdict(lambda: defaultdict(float))

for y in [n for n, d in G.nodes(data=True) if d.get('type') == 'Y']:
    weight = y_importance.get(y, 1.0)  # default to 1 if no Zs
    x_neighbors = [n for n in G.neighbors(y) if G.nodes[n].get('type') == 'X']
    t_neighbors = [n for n in G.neighbors(y) if G.nodes[n].get('type') == 'T']

    for x in x_neighbors:
        for t in t_neighbors:
            weighted_path_count[x][t] += weight
```
