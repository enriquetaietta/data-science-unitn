```python
import networkx as nx
import pandas as pd

```

```python
nodes = pd.read_csv('nodes_v3.csv')
edges = pd.read_csv('edges_v3.csv')
```

```python
nodes
```

```python
edges
```

```python
# Creazione di un grafo di esempio
G = nx.Graph()
for index, row in nodes.iterrows():
  G.add_node(row[0], type=row[1])

for index, row in edges.iterrows():
  G.add_edge(row[0], row[1], type=row[2])
```

```python

print('Number of nodes: ' + str(G.number_of_nodes()))
print('Number of edges: ' + str(G.number_of_edges()))
# Network-level metrics
density = nx.density(G)
print(f"Density: {density}")

## # Node-level metrics
## degree_centrality = nx.degree_centrality(G)
## betweenness_centrality = nx.betweenness_centrality(G)
## closeness_centrality = nx.closeness_centrality(G)
##
## print("Degree Centrality:")
## for node, centrality in degree_centrality.items():
##     print(f"Node {node}: {centrality}")
##
## print("Betweenness Centrality:")
## for node, centrality in betweenness_centrality.items():
##     print(f"Node {node}: {centrality}")
##
## print("Closeness Centrality:")
## for node, centrality in closeness_centrality.items():
##     print(f"Node {node}: {centrality}")

# Edge-level metrics
node_type = 'Y'

# Filter nodes by type
nodes_of_type = [node for node, attrs in G.nodes(data=True) if attrs.get('type') == node_type]
print("Numero di nodi di tipo Y: " + str(len(nodes_of_type)))

# Calculate the clustering coefficient for nodes of the specified type
clustering_coefficients = nx.clustering(G, nodes=nodes_of_type)

nodes_degree = nx.degree(G, nbunch=nodes_of_type)
print(nodes_degree)
print(type(nodes_degree))
for node, degree in nodes_degree:
  print(node)
  print(degree)

nodes_degree_list = list(nodes_degree)

print(nodes_degree_list[0])
print(nodes_degree_list[1])

nodes_degree_list.sort(key=lambda x: x[1], reverse=True)
print(nodes_degree_list[0])
print(nodes_degree_list[1])

##
# Print the clustering coefficients
for node, coefficient in clustering_coefficients.items():
    if (coefficient > 0):
        print(f"Node {node}: Clustering Coefficient = {coefficient}")
```
